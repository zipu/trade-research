```python
%matplotlib inline
import sys
sys.path.append('../')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from modules.backtesting import Market, Trader
from modules.tools import open_file, product_info
```

```python
feed = open_file('h5py', 'etc/ohlct.h5', mode='r')
```

```python
def gen_signal(datatable):
    #추세지속일수
    days = 3
    
    Market.set_ATR(datatable, span=20)
    
    #추세 누적 지속일수
    down = up = neutral= 0
    for date, metric in datatable.iterrows():
        trend = metric['trend']
        if trend == 0:
            down += 1
            up = 0
            neutral = 0
        elif trend == 1:
            neutral += 1
            up = 0
            down = 0 
        elif trend == 2:
            up += 1
            down = 0
            neutral = 0
        datatable.loc[date, 'up'] = up
        datatable.loc[date, 'down'] = down
        datatable.loc[date, 'neutral'] = neutral
```

```python
market = Market(feed=feed, signal=gen_signal)
```

```python
def strat(trader, info, metric):
    date = metric.name
    symbol = info['group']
    #entrylong
    if metric['up'] == 4:
        #if not trader.trades.isopen(info, Market.long, 'TF'):
        entryprice =  Market.get_price(info, metric['open'], metric['high'], skid=0.5)
        position = Market.long
        stopprice = entryprice - metric['ATR']*3
        trader.buy(info, 'TF', position, date, entryprice, stopprice)
    
    #entryshort
    if metric['down'] == 4:
        #if not trader.trades.isopen(info, Market.short, 'TF'):
        entryprice =  Market.get_price(info, metric['open'], metric['low'], skid=0.5)
        position = Market.short
        stopprice = entryprice + metric['ATR']*3
        #stopprice = trader.stop_rule(info, position, metric, entryprice)
        trader.buy(info, 'TF', position, date, entryprice, stopprice)
    
    #exit
    if (metric['neutral'] == 3):
        #exitlong
        trade = trader.trades.get(info,'TF',Market.long)
        if trade:
            exitprice = Market.get_price(info, metric['open'], metric['low'], skid=0.5)
            trader.sell(trade, date, exitprice)
            
        #exitshort
        trade = trader.trades.get(info,'TF',Market.short)
        if trade:
            exitprice = Market.get_price(info, metric['open'], metric['high'], skid=0.5)
            trader.sell(trade, date, exitprice)
```

```python
def stop_rule(self, inst, position, metric, entryprice):
    #초기 설정
    stopprice = entryprice - (position * metric['ATR'] * 3)
    if np.isnan(stopprice):
        print(metric)
    return stopprice
```

```python
%%time
trader = Trader(market=market,
                principal=100000,
                portfolio_heat=0.2,
                heat=0.05,
                max_lot= None,
                strategy=strat,
                stop=None)
trader.run_trade(start='2006-01-01', end='2017-01-01')
```
