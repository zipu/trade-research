### 추세 지속 일수 분석

```python
%matplotlib inline
import sys
from collections import OrderedDict
sys.path.append('../..')
import pandas as pd
import numpy as np

from modules.db_manager import open_file
from modules.backtesting import simulator
pd.options.display.float_format = '{:,.2f}'.format #화면에 소숫점 2번째 자리까지만 표시
```

```python
raw_file = open_file("h5py""40day_trend_labeled_futures.h5", mode='r')
```

```python
df = pd.DataFrame()
for idx, file in enumerate(raw_file.values()):
    #if idx > 10:
    #    break
        
    data = pd.DataFrame(file.value[:,[1,4,7]], index=file.value[:,0].astype('M8[s]'), columns=['open','close','trend'])
    #추세 누적 지속일수
    data['signal'] = (data.trend.diff(1) != 0).astype('int').cumsum()
    #추세 데이터 타입 변경
    data['trend'] = data.trend.replace(to_replace=[0.,1.,2.], value=['down','neutral','up'])
    grouped = pd.DataFrame({
        'days' : data.groupby('signal').size(),
        'trend': data.groupby('signal').trend.first()
    })
    df2 = grouped.groupby(['days', 'trend']).days.count().to_frame()
    df2.columns=[file.attrs['name']]
    df = pd.concat([df, df2], axis=1)
```

```python
df.insert(0, 'count', df.sum(axis=1))
```

```python
up = df.xs('up', level='trend')
up = up.reindex(index=np.arange(1, up.index.max()+1))
down = df.xs('down', level='trend')
down = down.reindex(index=np.arange(1, down.index.max()+1))
neutral = df.xs('neutral', level='trend')
neutral = neutral.reindex(index=np.arange(1, neutral.index.max()+1))
```

```python
# 누적 연속일수
ref = neutral
a = [ ref['count'].sum() - ref['count'].shift(i+1).sum() for i in range(len(ref))]
a[-1] = ref['count'].sum()
a.sort(reverse=True)
ref.insert(1,'cummulative', a)
ref.insert(2, 'ratio(%)', 100*ref['cummulative']/ref['cummulative'].max())
change = 100*ref['cummulative']/ref['cummulative'].shift(1)
change.iloc[0] = 100
ref.insert(3, 'change(%)', change)
```

```python
columns = [np.array(['uptrend','uptrend','uptrend','uptrend','downtrend','downtrend','downtrend','downtrend','neutral','neutral','neutral','neutral']), 
           np.array(['count','cummulative','ratio(%)','change(%)','count','cummulative','ratio(%)','change(%)','count','cummulative','ratio(%)','change(%)'])]
result = pd.concat([up[['count','cummulative','ratio(%)','change(%)']],down[['count','cummulative','ratio(%)','change(%)']], neutral[['count','cummulative','ratio(%)','change(%)']] ],
                    axis=1)
result.columns = columns
```

```python
result
```

```python
#각각 파일로 저장
up.to_csv('uptrend.csv')
down.to_csv('downtrend.csv')
neutral.to_csv('neutral.csv')
result.to_csv('summary.csv')
```
