```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output
import time
```

```python
symbol = 'JY'
pinfo = load_products()[symbol]
ohlc_raw = open_file('h5py', 'raw data/daily_futures.h5', mode='r')
tdop_raw = open_file('h5py', f'tdops/tdops_deday_7.h5', mode='r')

#tdop_db = pdens[symbol]
#tdop_db2 = pdens2[symbol]
name = list(filter(lambda x: x.attrs['symbol']==symbol, ohlc_raw.values()))[0].name
ohlc = dataframe(name, ohlc_raw)

tdops = tdop_raw[symbol]['tdop'].value
dates = tdop_raw[symbol]['dates'].value.astype('M8[s]')
price = tdop_raw[symbol]['prices'].value
```

```python
th = 0.99
bottom = []
top = []
for idx, date in enumerate(dates):
    print(f"\r {date}   ", flush=True, end='')
    tdop = tdops[idx]/tdops[idx].sum()
    t = tdop
    tdop = tdop.cumsum()
    cond = np.where( (tdop>1-th) & (tdop< th) )[0]
    
    if cond.size > 0:
        bottom.append(price[cond.min()])
        top.append(price[cond.max()])
    else:
        bottom.append(np.nan)
        top.append(np.nan)
```

```python
data = ohlc[:]
fig=plt.figure(figsize=(15,9))
ax=fig.add_subplot(111, label="ohlc")
ax = ohlc_chart(ax, data)
ax.plot(data.index, bottom, color='blue')
ax.plot(data.index, top, color='blue')


start = np.datetime64('2012-01-20')
end = np.datetime64('2014-04-05')
ax.set_xlim([start,end])

min = ohlc.loc[start:end].low.min()*0.95
max = ohlc.loc[start:end].high.max()*1.02
ax.set_ylim([min, max])
```
