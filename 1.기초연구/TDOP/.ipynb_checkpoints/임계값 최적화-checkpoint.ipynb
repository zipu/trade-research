```python
bottom = []
top = []
for idx, date in enumerate(dates):
    print(f"\r {date}   ", flush=True, end='')
    
    tdop = tdops[idx]/tdops[idx].sum()
    #tdop = norm(tdop, ntype='minmax')
    
    for th in np.arange(0.00001,0.01, 0.00001):
        if tdop[tdop>th].sum() < 0.99:
            break
    
    cond = np.where(tdop>th)[0]
    bottom.append(cond.min())
    top.append(cond.max())
```

```python
data = ohlc[:]
fig=plt.figure(figsize=(15,9))
ax=fig.add_subplot(111, label="ohlc")
ax = ohlc_chart(ax, data)
ax.plot(data.index, price[bottom], color='blue')
ax.plot(data.index, price[top], color='blue')


start = np.datetime64('2006-02-01')
end = np.datetime64('2010-09-01')
ax.set_xlim([start,end])

min = ohlc.loc[start:end].low.min()*0.95
max = ohlc.loc[start:end].high.max()*1.02
ax.set_ylim([min, max])
```
