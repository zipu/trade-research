```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tables as tb
from datetime import datetime
np.random.seed(777)
```

```python
db = tb.open_file("data/market.hdf5", mode='r')
```

```python
# 종목코드
group = 'URO'
```

```python
# 일일 ohlc 데이터프레임 생성
daily = pd.DataFrame(getattr(db.root, group).Daily.read())
daily = daily.sort_values('date', ascending=True) # 날짜 정렬
daily = daily.assign(date=daily.date.astype('M8[s]'))

daily['change'] = daily.close.pct_change()
acor = daily.sort_values('date', ascending=False).change[:40]
```

```python
df = daily.sort_values('date', ascending=False)
df = df.reset_index(drop=True)
timeframe = 10
cor = []
dates = []
for i in range(len(df) - timeframe):
    ret = df.change[i:i+timeframe].autocorr()
    cor.append(ret)
    dates.append(df.date[i])
```

```python
ax = plt.plot(dates, cor)
ax = plt.gca()
ax.figure.set_size_inches(10,8)
ax.set_xlim([datetime(2017,3,3,1,1,1),datetime.today()])
```
