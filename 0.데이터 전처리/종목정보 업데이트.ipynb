### Carpediem project의 db파일로부터 종목정보를 갱신

```python
import sys
sys.path.append('../')
import sqlite3 as lite
import os
import pandas as pd
```

```python
# Ebest products info
fpath = os.path.join('../../carpediem/data/', 'db.sqlite3')
con = lite.connect(fpath)
products = pd.read_sql('select * from trading_product', con)
products.set_index(['group'], drop=False, inplace=True)
products = products[products.market != 'ETC']
```

```python
# Quandl products info
df = pd.read_csv('../data/SCF-meta.csv')
```

```python
df.set_index('Ebest Symbol', inplace=True)
```

```python
pinfo = products[['name','currency','keep_margin','tick_unit','tick_value','decimal_places']].copy(deep=True)
pinfo['symbol'] = products['group']
df2 = pd.concat([pinfo, df['sector']], axis=1, join='inner')
df2.to_json('../data/product_info.json', orient='index')
```
